```python
# q2
def crop_grid(img, num_horizontal_grid, num_vertical_grid, line_color):
    # Get image dimensions
    height, width = img.shape[:2]
    
    # Calculate the width and height of each grid cell
    M = height // num_vertical_grid
    N = width // num_horizontal_grid

    # Create a copy of the image to draw the grid on
    img_copy = img.copy()

    for y in range(0, height, M):
        for x in range(0, width, N):
            y1 = y + M
            x1 = x + N

            if x1 >= width and y1 >= height:
                x1 = width - 1
                y1 = height - 1
            elif y1 >= height:
                y1 = height - 1
            elif x1 >= width:
                x1 = width - 1

            cv.rectangle(img_copy, (x, y), (x1, y1), line_color, 1)

    return img_copy

# Example usage:
img = cv.imread('images/11.jpg')
num_horizontal_grid = 3
num_vertical_grid = 2
line_color = (0, 255, 0)  # Green color

# Get the image with grid lines
img_with_grid = crop_grid(img, num_horizontal_grid, num_vertical_grid, line_color)

# Display the image with grid lines
cv.imshow('Patched Image', img_with_grid)
cv.waitKey(0)
cv.destroyAllWindows()

```

```python
def adjust_brightness_opencv(image, value):
    hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)
    h, s, v = cv.split(hsv)

    v = cv.add(v, value)
    v = np.clip(v, 0, 255)

    final_hsv = cv.merge((h, s, v))
    img_bright = cv.cvtColor(final_hsv, cv.COLOR_HSV2BGR)
    
    return img_bright

# Load image
img_path = 'images/11.jpg'  # Ensure this path is correct
image = cv.imread(img_path)

# Adjust brightness
value = 50  # Increase brightness by 50
bright_img = adjust_brightness_opencv(image, value)

# Display the images
plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv.cvtColor(bright_img, cv.COLOR_BGR2RGB))
plt.title('Brightness Adjusted Image')
plt.axis('off')

plt.show()
```

```python
def adjust_brightness_numpy(image, value):
    img_bright = np.clip(image + value, 0, 255).astype(np.uint8)
    
    return img_bright

# Load image
img_path = 'images/11.jpg'  # Ensure this path is correct
image = cv.imread(img_path)

# Adjust brightness
value = 50  # Increase brightness by 50
bright_img = adjust_brightness_numpy(image, value)

# Display the images
plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv.cvtColor(bright_img, cv.COLOR_BGR2RGB))
plt.title('Brightness Adjusted Image')
plt.axis('off')

plt.show()
```
